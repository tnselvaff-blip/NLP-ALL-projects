```python
import pandas as pd

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Sample dataset
data = pd.DataFrame(
    {
        'Movie': ['Movie A', 'Movie B', 'Movie C'],
        'Description': [
            'Action and adventure in space',
            'Romantic comedy with friends',
            'Sci-fi thriller in space'
        ]
    }
)
```

```python
# TF-IDF vectorization
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(data['Description'])

# Cosine similarity
cos_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Recommend movies similar to Movie A
movie_idx = 0
similar_movies = list(enumerate(cos_sim[movie_idx]))
similar_movies = sorted(similar_movies, key=lambda x: x[1], reverse=True)

print("Recommendations for Movie A:")
```

```python
# Recommend movies similar to Movie A

movie_idx = 0

similar_movies = list(enumerate(cos_sim[movie_idx]))

similar_movies = sorted(similar_movies, key=lambda x: x[1], reverse=True)

print("Recommendations for Movie A:")

for idx, score in similar_movies[1:]:
    print(data['Movie'][idx], "Score:", score)
```
