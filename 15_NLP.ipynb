```python
import nltk # Corrected from nitk

from nltk.corpus import stopwords

from nltk.tokenize import word_tokenize, sent_tokenize

import heapq

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab') # Added to download the missing resource

text = """Artificial Intelligence (AI) is a branch of computer science that aims to create intelligent machines.
It has applications in various fields such as healthcare, finance, education, and autonomous vehicles. Machine Learning, a subset of AI, allows systems to learn from data without explicit programming.
AI continues to grow rapidly and impacts everyday life in multiple ways.
Natural Language Processing enables machines to understand and interpret human language.""" # Corrected text assignment

stop_words = set(stopwords.words("english"))

words = word_tokenize(text.lower()) # Added assignment operator '='

word_frequencies = {} # Corrected dictionary initialization

for word in words:
    if word not in stop_words and word.isalpha():
        word_frequencies[word] = word_frequencies.get(word, 0) + 1 # Corrected default value from 8 to 0

sentences = sent_tokenize(text)

sentence_scores = {} # Corrected dictionary initialization

for sent in sentences:
    for word in word_tokenize(sent.lower()):
        if word in word_frequencies:
            if sent not in sentence_scores:
                sentence_scores[sent] = 0
            sentence_scores[sent] += word_frequencies[word]
```
